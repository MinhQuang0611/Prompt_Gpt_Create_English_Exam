```python
from spellchecker import SpellChecker
spell = SpellChecker()
```

```python
spell['morning']
'morning' in spell
spell.known(['morning', 'hapenning'])
spell.unknown(['morning', 'hapenning'])
```

```python
words = ['morning', 'stpid']
# Nhận diện từ có đúng chính tả hay không, nếu có thêm vào valid word
valid_word = spell.known(words)
print(valid_word)
print()
#In ra từ đúng chính tả
corrected_words = {word: spell.correction(word) for word in words}
print(corrected_words)
```

```python
# Thêm và xóa các thuật ngữ khỏi từ điển
spell = SpellChecker()
#Đẩy file json
# spell.word_frequency.load_dictionary("path.json") # Em không có file nên để tượng trunngw thôi hehe :)
# file txt
# spell.word_frequency.load_text_file("path.txt")
# Text thuong
spell.word_frequency.load_text("You are the most stupid in the world")
#  Thêm nhiều từ
spell.word_frequency.load_words(['i', 'am', 'smart'])
# Thêm 1 từ 
spell.word_frequency.add("victory")
#Xóa từ
spell.word_frequency.remove_words(['i', 'am'])
spell.word_frequency.remove("victory")

for word in spell:
    print("{}: {}".format(word, spell[word]))
                                  
```
